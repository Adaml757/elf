```python
from elfpy import elf_io_readwrite, elf_io_correctdng
import numpy as np
from jupyter_helpers import *

# Load parameter and infosum structure
fname = "G:\data\ELF dev tests\\245 Adaminaby grass plains\mat\\245 Adaminaby grass plains_info.json"
para = elf_io_readwrite([], "loadpara", fname)
infosum = elf_io_readwrite(para, "loadinfosum")
print(para)
print(infosum)
```

```python
# Load a scene image from its tif (uncalibrated)
im = elf_io_readwrite(para, 'loadHDR_tif', 'scene001')
display_img(im, figsize=(10,10))
```

```python
# Load a scene image from its mat (calibrated)
im = elf_io_readwrite(para, "loadHDR_mat", "scene001")
im_corr = elf_io_correctdng(im, infosum, "default")
        
display_img(im_corr, figsize=(10,10), grid=infosum["grids"]["scene"]["fisheye"])
```

```python
# Load a filtered image from its mat (calibrated)
im = elf_io_readwrite(para, "loadfilt_mat", "scene001")
im_corr = elf_io_correctdng(im[0], infosum, "bright")
display_img(im_corr, figsize=(10,10), grid=infosum["grids"]["filter"][0]["fisheye"])
im_corr = elf_io_correctdng(im[1], infosum, "bright")
display_img(im_corr, figsize=(10,10), grid=infosum["grids"]["filter"][1]["fisheye"])
```

```python
# Load all scene images for an environment from mats (calibrated)
im_stack = np.zeros((7,2000, 2000, 3), dtype='float')
for setnr in range(1, 8):
    im = elf_io_readwrite(para, "loadHDR_mat", f'scene{setnr:03d}')
    im_corr = elf_io_correctdng(im, infosum, "bright")  # we could use the "maxval" method here to normalise all images to the same value 
    im_stack[setnr-1,:,:,:] = np.expand_dims(im_corr, 0)
display_img_stack(im_stack)
```
